```python
# This function does not print
def deal(): 
    deck = list(range(1,14))*4
    np.random.shuffle(deck)
    player1 = tuple(deck[:int(len(deck)/2)]) # Stores 26 random values in a tuple
    player2 = tuple(deck[int(len(deck)/2):]) # Stores 26 random values in a tuple
    return player1, player2
```

### Step 1.1 and Step 1.2
#### Add the capability to track the number of ties for each game
#### Add the capability to track the largest total won during a round in a given game

```python
# This function does not print
def play(): 
    num_of_cards = 52
    num_of_cards_per_hand = 26
    total_rounds = 13
    player1_total = 0
    player2_total = 0
    record = 0
    no_of_ties = 0 # Used to track number of ties (Step 1.1)
    largest_round_total = 0 # Used to track number of the largest round total (Step 1.2)
    player1,player2 = deal() 
    for i in range(total_rounds):
        if player1[i] + player1[i + total_rounds] > player2[i] + player2[i + total_rounds]:
            player1_total += (player1[i] + player1[i + total_rounds] + player2[i] + player2[i + total_rounds] + record)
            
            # Update the largest total win
            if largest_round_total < (player1[i]+player1[i+total_rounds]+player2[i]+player2[i+total_rounds]+ record):
                largest_round_total = (player1[i]+player1[i+total_rounds]+player2[i]+player2[i+total_rounds]+ record)
            record = 0

        elif player1[i]+player1[i+total_rounds] < player2[i]+player2[i+total_rounds]:
            player2_total += (player1[i]+player1[i+total_rounds]+player2[i]+player2[i+total_rounds]+ record)
            
            # Update the largest total win
            if largest_round_total < (player1[i]+player1[i+total_rounds]+player2[i]+player2[i+total_rounds]+ record):
                largest_round_total = (player1[i]+player1[i+total_rounds]+player2[i]+player2[i+total_rounds]+ record)
            record = 0
        else:
            record = (player1[i]+player1[i+total_rounds]+player2[i]+player2[i+total_rounds]+ record)
            if (i == total_rounds-1):
                player1_total += record/2
                player2_total += record/2
            no_of_ties += 1
            
    store = {}
    store['player1_total'] = player1_total
    store['player2_total'] = player2_total
    store['player1_initial'] = sum(player1) # Sum of initial cards dealt to player1
    store['player2_initial'] = sum(player2) # Sum of initial cards dealt to player2
    store['no_of_ties'] = no_of_ties # Count the number of ties in a game
    store['largest_round_total'] = largest_round_total # Store the largest total won in a game
        
    if player1_total > player2_total:
        store['final_winner'] = 1
        store['player1_final'] = True
        store['player2_final'] = False
        store['final_tie'] = False
        store['winner_initial'] = sum(player1)
        store['winner_total'] = player1_total
        store['loser_initial'] = sum(player2)
        store['loser_total'] = player2_total
        return store
    elif player1_total < player2_total:
        store['final_winner'] = 2
        store['player1_final'] = False
        store['player2_final'] = True
        store['final_tie'] = False
        store['winner_initial'] = sum(player2)
        store['winner_total'] = player2_total
        store['loser_initial'] = sum(player1)
        store['loser_total'] = player1_total
        return store
    else:
        store['final_winner'] = 0
        store['player1_final'] = False
        store['player2_final'] = False
        store['final_tie'] = True
        store['winner_initial'] = (sum(player1)+sum(player2))/2
        store['winner_total'] = (player1_total+player2_total)/2
        store['loser_initial'] = (sum(player1)+sum(player2))/2
        store['loser_total'] = (player1_total+player2_total)/2
        return store
```

### Step 1.3
#### The function should track the following statistics for each game, stored in a dictionary (ideally as separate entries, which will facilitate next steps):

#### 1. Initial totals (i.e., sum of card values for initially dealt hands) for player 1 and player 2, respectively
#### 2. Number of ties during gameplay
#### 3. Maximum round total won (i.e., maximum sum of card values won during a round, potentially including tiebreakers)
#### 4. Final totals (i.e., sum of card values after completing gameplay) for player 1 and player 2, respectively

```python
def play_new (games = 1): # default values for games = 1
    dict1 = {}
    dict1['player1_initial'] = []
    dict1['player2_initial'] = []
    dict1['no_of_ties'] = []
    dict1['largest_round_total'] = []
    dict1['player1_total'] = []
    dict1['player2_total'] = []
    for y in range(games):
        x = play()
        dict1['player1_initial'].append(x['player1_initial'])
        dict1['player2_initial'].append(x['player2_initial'])
        dict1['no_of_ties'].append(x['no_of_ties'])
        dict1['largest_round_total'].append(x['largest_round_total'])
        dict1['player1_total'].append(x['player1_total'])
        dict1['player2_total'].append(x['player2_total'])
    return dict1
```

### Step 3.1
#### Add a new column to your DataFrame ('outcome') that contains the value 'Player 1 wins' if Player 1's final total is higher, 'Player 2 wins' if Player 2's final total is higher, or 'Tie game' if the two player's final totals are equal

```python
def outcome(player1,player2):
    if player1 > player2:
        return 'Player 1 wins'
    elif player1 < player2:
        return 'Player 2 wins'
    else:
        return "Tie game"

df['outcome'] = df.apply(lambda x : outcome(x['player1_total'], x['player2_total']), axis=1)
```

### Step 3.2

#### Add new columns for the winning total ('winning_total') and losing total ('losing_total'), which contain the final totals for the winning and losing players, respectively, and missing values (NumPy.nan) for tie games

```python
def winning_total(player1,player2):
    if player1 > player2:
        return player1
    elif player1 < player2:
        return player2
    else:
        return np.nan

def losing_total(player1,player2):
    if player1 > player2:
        return player2
    elif player1 < player2:
        return player1
    else:
        return np.nan
df['winning_total'] = df.apply(lambda x: winning_total(x['player1_total'],x['player2_total']),axis=1)
df['losing_total'] = df.apply(lambda x: losing_total(x['player1_total'],x['player2_total']),axis=1)
```
